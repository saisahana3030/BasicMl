# IMPORTANT LIBRARIES


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# READ THE DATASET('TITANIC')

```python
log=pd.read_csv('titanic_train.csv')
log.head(8)
```

# TO KNOW THE VALUES ARE NULL OR NOT

```python
log.isnull()
```

# To Know What Datatype It Is

```python
log.Name.dtype
```

```python
log.Sex.dtype
```

# LINE PLOT

```python
sns.lineplot(data=log,x="Age",y="Sex")
plt.show()
```

# COUNT PLOT

```python
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Sex',data=log,palette='BuGn')
```

```python
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Pclass',data=log,palette='cubehelix')
```

```python
sns.countplot(x='SibSp',data=log,palette='Blues_r')
```

# BAR PLOT

```python
sns.barplot(data=log,x='Pclass',y="Survived")
```

# DISTPLOT

```python
sns.distplot(log['Age'].dropna(),color='green',bins=10)
```

# BOX-PLOT

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='Pclass',y='Age',data=log,palette='Dark2_r')
```

# SCATTER PLOT

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Pclass',y='Age',data=log,palette='Dark2_r')
```

```python
sns.pairplot(log)
```

# HEAT-MAP-(NULL VALUES)

```python
sns.heatmap(log.isnull())
```

# CLEANING DATASET

```python
log.drop('Cabin',axis=1,inplace=True)
```
